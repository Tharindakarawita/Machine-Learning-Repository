```python
# rename column_names
new_colunms_name = []
for i in list(df_ohe.columns):
    split = i.split('_')

    if i == 'class':
        new_colunms_name.append('class')
        continue
    # print(split[0])
    elif split[0] in mushroom_info:
        longname_string = [key for key, value in mushroom_info[split[0]].items() if value == split[1]][0]
        combine_string = split[0]+'_'+longname_string
        new_colunms_name.append(combine_string)

rename_colunm = {}
for n,string in enumerate(list(df_ohe.columns)):
    rename_colunm[string] = new_colunms_name[n]

rename_colunm
df_ohe = df_ohe.rename(columns=rename_colunm)
print(df_ohe.columns)
```

```python

from sklearn.metrics import confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
# Assuming X_test, y_test, and classifier are defined
# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()

classes = ['Edible', 'Poisonous']
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)

for i in range(len(classes)):
    for j in range(len(classes)):
        plt.text(j, i, str(conf_matrix[i, j]), ha='center', va='center', color='white')

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, classifier.predict_proba(X_test)[:, 1])
roc_auc = auc(fpr, tpr)

# Plot ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

# Confustion Matricx

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

def plot_confusion_matrix(ax, model, X, y, title):
    model.fit(X, y)

    # Make predictions on the test set
    y_pred = model.predict(X)

    # Confusion Matrix
    conf_matrix = confusion_matrix(y, y_pred)

    # Plot Confusion Matrix
    im = ax.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    ax.set_title(title)
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])

    classes = ['Edible', 'Poisonous']
    tick_marks = range(len(classes))
    ax.set_xticklabels(classes)
    ax.set_yticklabels(classes)

    for i in range(len(classes)):
        for j in range(len(classes)):
            ax.text(j, i, str(conf_matrix[i, j]), ha='center', va='center', color='red')

    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

    return im

# Model instances
log_model = LogisticRegression()
svm_model = SVC(C=1.0, kernel='rbf')
MLP_model = MLPClassifier(hidden_layer_sizes=(128, 128))
ada_model = AdaBoostClassifier()
naive_bayes_model = GaussianNB()
decision_tree_model = DecisionTreeClassifier()

# Plotting Confusion Matrices
fig, axes = plt.subplots(3, 6, figsize=(18, 9))

# Confusion matrices for the training set
plot_confusion_matrix(axes[0, 0], log_model, X_train, y_train, 'Logistic Regression (Training Set)')
plot_confusion_matrix(axes[0, 1], svm_model, X_train, y_train, 'Support Vector Machine (Training Set)')
plot_confusion_matrix(axes[0, 2], MLP_model, X_train, y_train, 'MLPClassifier (Training Set)')
plot_confusion_matrix(axes[0, 3], ada_model, X_train, y_train, 'AdaBoost (Training Set)')
plot_confusion_matrix(axes[0, 4], naive_bayes_model, X_train, y_train, 'Naive Bayes (Training Set)')
plot_confusion_matrix(axes[0, 5], decision_tree_model, X_train, y_train, 'Decision Tree (Training Set)')

# Confusion matrices for the testing set
plot_confusion_matrix(axes[1, 0], log_model, X_test, y_test, 'Logistic Regression (Testing Set)')
plot_confusion_matrix(axes[1, 1], svm_model, X_test, y_test, 'Support Vector Machine (Testing Set)')
plot_confusion_matrix(axes[1, 2], MLP_model, X_test, y_test, 'MLPClassifier (Testing Set)')
plot_confusion_matrix(axes[1, 3], ada_model, X_test, y_test, 'AdaBoost (Testing Set)')
plot_confusion_matrix(axes[1, 4], naive_bayes_model, X_test, y_test, 'Naive Bayes (Testing Set)')
plot_confusion_matrix(axes[1, 5], decision_tree_model, X_test, y_test, 'Decision Tree (Testing Set)')

# Confusion matrices for the basic set
plot_confusion_matrix(axes[2, 0], log_model, X_basic, Y_basic, 'Logistic Regression (Basic Set)')
plot_confusion_matrix(axes[2, 1], svm_model, X_basic, Y_basic, 'Support Vector Machine (Basic Set)')
plot_confusion_matrix(axes[2, 2], MLP_model, X_basic, Y_basic, 'MLPClassifier (Basic Set)')
plot_confusion_matrix(axes[2, 3], ada_model, X_basic, Y_basic, 'AdaBoost (Basic Set)')
plot_confusion_matrix(axes[2, 4], naive_bayes_model, X_basic, Y_basic, 'Naive Bayes (Basic Set)')
plot_confusion_matrix(axes[2, 5], decision_tree_model, X_basic, Y_basic, 'Decision Tree (Basic Set)')

fig.tight_layout()
plt.show()
```
